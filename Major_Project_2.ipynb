```python
#importing necessary libraries 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#9- determining minimum and maximum number of likes
print(np.min(df['likes']))
print(np.max(df['likes']))
```

```python
#10- distribution plot for comments column
plt.figure(figsize= (10,5))
sns.distplot(df['comments'])
```

```python
#11- barplot between likes and views
x= df.iloc[:20, :]
plt.figure(figsize= (20,5))
sns.barplot(x=x['likes'], y=x['views'])
```

```python
#12- plot between likes and comments
plt.figure(figsize= (10,5))
plt.plot(x['likes'], x['comments'])
```

```python
#13- scatter plot between likes and comments 
plt.scatter(x['likes'], x['comments'], c='r', s= 20, marker= 'd')
```
